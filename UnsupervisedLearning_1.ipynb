```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# load data
data = pd.read_csv(r'C:\Users\HP 840G4\AppData\Local\Temp\Temp922d94b3-52b1-49b3-945e-bec4712f9b24_datasets.zip\archive (2).zip')
data
```

```python
# Drop the CUST_ID column and handle missing data
data = data.drop("CUST_ID", axis=1)
data = data.fillna(data.median())
```

```python
# scaling the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Determinig the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# Apply K-means clustering with the optimal number of clusters (k=4)
k = 4
kmeans = KMeans(n_clusters=k, init="k-means++", random_state=42)
clusters = kmeans.fit_predict(data_scaled)
```

```python
import seaborn as sns

# Add the cluster labels to the original dataset
data["Cluster"] = clusters

# Visualize the clusters using a pairplot with two selected features (e.g., BALANCE and PURCHASES)
sns.pairplot(data, vars=["BALANCE", "PURCHASES"], hue="Cluster")
plt.show()

# Analyze the characteristics of each cluster
cluster_summary = data.groupby("Cluster").mean()
print(cluster_summary)
```
