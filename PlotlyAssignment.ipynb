```python
import pandas as pd
import plotly.express as px
```

```python
taitanic =  'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(taitanic)
data.head()
```

```python
data.info()
```

```python
na_df = data[data.isna().any(axis=1)]
na_df
```

```python
data = data.fillna(0)
data.head(15)
```

#  A bar plot to visualize the number of passengers who survived and did not survive, broken down by gender.

```python
# import plotly
import plotly.express as px
```

```python
# bar plot
barplot = px.bar(data, x='Sex', y='Survived', color='Sex', title='servival rate by gender')
barplot.update_xaxes(title='Gender')# modify the x lable
barplot.update_yaxes(title='number of servivers')# modify the y lable
barplot.show()
```

# A histogram to display the distribution of passengers' ages.

```python
data.head(3)
          
          
           
```

```python
histo = px.histogram(data, x='Age', nbins=20, title='servivers ages', color='Survived')
histo.update_yaxes(gridcolor='grey', title='servivers')
histo.show()
```

#  A violin plot to visualize the distribution of fare prices paid by passengers, broken down by their class (1st, 2nd, and 3rd class).



```python
data['Pclass'].value_counts()
```

```python
violinplot = px.violin(data, x='Pclass', y='Fare', color='Pclass', title='prices in relation to classes')
violinplot.show()
```

# Scatter plot to explore the relationship between passanger 'ages' and 'fare' prices

```python
data.info()
```

```python
import plotly.express as px

scatter = px.scatter(data,
                     x='Age',
                     y='Fare',
                     color='Pclass',
                     title='relationship between ages and fare prices',
                     symbol='Pclass')
scatter.update_layout(title=dict(text='relationship between ages and fare prices', xanchor='left'))
scatter.show()
```

# Using faceting to create a scatter plot matrix displaying the relationship between age, fare, and class for passengers who survived and did not survive.

```python
fig = px.scatter(data,
                     x='Age',
                     y='Fare',
                     color='Survived',
                     facet_col='Pclass',
                     title='relationship of age and fare in diffrent classes')
fig.show()
```
