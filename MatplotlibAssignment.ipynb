```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\HP 840G4\Desktop\data sci\datasets.zip')
```

# line plot showing monthly revenue over the two year period

```python
plt.figure(figsize=(10, 7))
plt.plot(df['Month'], df['Revenue'], color='blue', linestyle='-',linewidth=2, label = 'revenue')
plt.xlabel("months")
plt.ylabel("revenue")
plt.title("monthly revenue")

plt.xticks (rotation=45)
plt.legend()
plt.grid(color = 'grey', linestyle = '--')
plt.show()
```

#  bar plot showing the total number of orders per month.

```python
plt.figure(figsize=(10,7))
plt.bar(df['Month'], df['Orders'], color='brown', label='orders')
plt.xlabel('months')
plt.ylabel('orders')
plt.title('number of orders')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# A scatter plot showing the relationship between the number of orders and the average order value.

```python
plt.figure(figsize=(10, 7))
plt.scatter(df['Orders'],df['Average_Order_Value'], color='orange')
plt.xlabel('orders')
plt.ylabel('avarage_order_value')
plt.title('relationship btw orders and avarage order')

plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r'C:\Users\HP 840G4\Desktop\data sci\datasets.zip')
fig,(ax1,ax2,ax3) = plt.subplots(nrows=3, ncols=1, figsize=(20,25))

# line plot showing the monthly reveue
ax1.plot(df['Month'],df['Revenue'],color='red',linestyle='-',linewidth=2, label='monthy revenue')
ax1.set_title('monthly revenue')
ax1.set_xlabel('months')
ax1.set_ylabel('revenue')
ax1.tick_params(axis='x', rotation=45)
ax1.legend()
ax1.grid(color='grey',linewidth=0.5,linestyle='--')

# create a bar plot showing the total number of oders per month
ax2.bar(df['Month'], df['Orders'],color='blue',label='orders')
ax2.set_title('orders per month')
ax2.set_xlabel('months')
ax2.set_ylabel('orders')
ax2.tick_params(axis='x',rotation=45)
ax2.legend()
ax2.grid(color='grey',linewidth=0.5,linestyle='--',axis='y')

# scatter plot of number of orders vs avrage order values
ax3.scatter(df['Orders'],df['Average_Order_Value'], color='orange')
ax3.set_title('number of orders vs avrage order values')
ax3.set_xlabel('orders')
ax3.set_ylabel('avrage order value')
ax3.grid()

#save the subplot
plt.tight_layout()
plt.savefig('sales_data_analysis.png', dpi=300)
plt.show()
```
